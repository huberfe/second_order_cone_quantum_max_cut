# n = 96

```julia
l = 4
h = 4
J = 1
n = 6*l*h
A_open = shuriken(l,h; J=J, periodic=false); 
A      = shuriken(l,h; J=J, periodic=true); 

val, X = QMC_SOC(A, verbose=false, pauli_NPA1=false, VarBench_norm=true)
plot = QMC_Plot(A_open,X)
name = "Shuriken_" * string(l) * string(h) * "_" * string(J) * ".pdf"
Compose.draw(Compose.PDF(name, 16Compose.cm, 16Compose.cm), plot) #print to file
print(n); print("  "); print(val); plot
```

## Compare with VarBench 
### $n=96$, Energy: $-168.2918$
### careful! they use different normalization, Eq S2 in  https://arxiv.org/abs/2302.04919
### $\sum_{(i,j) \in E; a=x,y,z}  \sigma_i^a \sigma_j^a$

```julia
# energy in VarBench normalization


QMC_to_varbench(A, val)
```
