```julia
include("QMC.jl")
include("QMC_plot.jl")
include("Lattice.jl")
include("pyrochlore_v2.jl")
```

__First the 2x2x2 periodic one__

```julia
A=pyro_unit_4(2)
mat_size=size(A)
##number of edges
ones(Int64, 1, mat_size[1])*A*ones(Int64, mat_size[1], 1)/2
```

Hence the trivial bound is __-3*96=-288__

```julia
QMC_SOC(A; pauli_NPA1=false, starbound=false, starbound4=false, SDP_form=false, verbose=false, VarBench_norm=true)
```

```julia
##now adding NPA1
QMC_SOC(A; pauli_NPA1=true, starbound=false, starbound4=false, SDP_form=false, verbose=false, VarBench_norm=true)
```

```julia
##adding things on 4 RDM?
QMC_SOC(A; pauli_NPA1=true, starbound=false, starbound4=true, SDP_form=false, verbose=false, VarBench_norm=true)
```
