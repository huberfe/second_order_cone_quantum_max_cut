```julia
include("QMC.jl");
include("QMC_plot.jl");
include("Lattice.jl");
```

```julia
pauli = true
print("Pauli level 1:", pauli, "\n")

#### Square
print("Square \n")

# n = 18
l = 4; h = 4; J = 1
n = l*h
A      = square(l,h; J=J, periodic=true);  
nE = sum(A)/2
val, X = QMC_SOC(A, pauli_NPA1=pauli)
println(n, ", ", nE, ":  ", QMC_to_varbench(A, val))

# n = 36
l = 6; h = 6; J = 1
n = l*h
A      = square(l,h; J=J, periodic=true);  
nE = sum(A)/2
val, X = QMC_SOC(A, pauli_NPA1=pauli)
println(n, ", ", nE, ":  ", QMC_to_varbench(A, val))

# n = 64
l = 8; h = 8; J = 1
n = l*h
A      = square(l,h; J=J, periodic=true);  
nE = sum(A)/2
val, X = QMC_SOC(A, pauli_NPA1=pauli)
println(n, ", ", nE, ":  ", QMC_to_varbench(A, val))

# n = 100
l = 10; h = 10; J = 1
n = l*h
A      = square(l,h; J=J, periodic=true);  
nE = sum(A)/2
#val, X = QMC_SOC(A, pauli_NPA1=pauli)
#println(n, ", ", nE, ":  ", QMC_to_varbench(A, val))


#### Kagome
print("Kagome \n")

# n = 18
l = 2; h = 3; J = 1
n = 3*l*h
A      = kagome(l,h; J=J, periodic=true);  
nE = sum(A)/2
val, X = QMC_SOC(A, pauli_NPA1=pauli)
println(n, ", ", nE, ":  ", QMC_to_varbench(A, val))

# n = 48
l = 4; h = 4; J = 1
n = 3*l*h
A      = kagome(l,h; J=J, periodic=true); 
nE = sum(A)/2
val, X = QMC_SOC(A, pauli_NPA1=pauli)
println(n, ", ", nE, ":  ", QMC_to_varbench(A, val))

# n = 75
l = 5; h = 5; J = 1
n = 3*l*h
A      = kagome(l,h; J=J, periodic=true); 
nE = sum(A)/2
val, X = QMC_SOC(A, pauli_NPA1=pauli)
println(n, ", ", nE, ":  ", QMC_to_varbench(A, val))


#### Shuriken
print("Shuriken \n")

# n = 24
l = 2; h = 2; J = 1
n = 6*l*h
A      = shuriken(l,h; J=J, periodic=true);
nE = sum(A)/2
val, X = QMC_SOC(A, pauli_NPA1=pauli) 
println(n, ", ", nE, ":  ", QMC_to_varbench(A, val))

n = 96
l = 4; h = 4; J = 1
n = 6*l*h
A      = shuriken(l,h; J=J, periodic=true); 
nE = sum(A)/2 
#val, X = QMC_SOC(A, pauli_NPA1=pauli) 
#println(n, ", ", nE, ":  ", QMC_to_varbench(A, val))
```
