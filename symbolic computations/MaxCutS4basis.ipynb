```python
# variable reduction for positivity of CS_4 elements
# lots of copy pasting output to define variable. (Substitution did not work due to type change!)

# key idea here: construct elements in group ring; replace by commutative variables;
# do substitutions; print new equations with newly intialized variables (initialized as permutations)

S4 = SymmetricGroup(4)
S  = SymmetricGroup(4).algebra(QQ)

def involution(a):
    """ involutes elements of group algebra f symmetric group 
        input: a, element of C[S_k]
        output: conjugate or involuted element in CS_m
    """
    # [el for el in X] returns the list tuples wiuth permutations and coefficients
    return sum([el[1].conjugate() * S(el[0].inverse()) for el in a])

def per_to_str(pi):
    pi_name =  str(pi)
    pi_name = pi_name.replace(')(', '_')
    pi_name = pi_name.replace(')', '')
    pi_name = pi_name.replace('(', '')
    pi_name = pi_name.replace(',', '')
    return pi_name
```

```python
# identities for S4 on four qubits
#id0_lhs = S((1,2,3)) 
#id0_rhs = S((1,2)) + S((2,3)) + S((1,3)) - S.one()  -  S((1,3,2))
#id1_lhs = S((1,2,3,4)) 
#id1_rhs = - S.one() + S((1,3)) + S((2,4)) \
#      + S([(1,2),(3,4)]) + S([(1,4),(2,3)]) - S([(1,3),(2,4)])  -  S((4,3,2,1))

#### a replacements
# maps also a_132 - > a_123  etc one representative for pi and pi^{-1}

# 3-cycle replacements
aid0_lhsA = a_123
aid0_rhsA = (a_12 + a_23 + a_13 - a_)/2

aid0_lhsB = a_124
aid0_rhsB = (a_12 + a_24 + a_14 - a_)/2

aid0_lhsC = a_134
aid0_rhsC = (a_13 + a_34 + a_14 - a_)/2

aid0_lhsD = a_234
aid0_rhsD = (a_23 + a_34 + a_24 - a_)/2

#s_1432, s_1234
aid1_lhsA = a_1234
aid1_rhsA = (- a_ + a_13 + a_24 \
      + a_12_34 + a_14_23 - a_13_24)/2

#s_1243, s_1342
aid1_lhsB = a_1243
aid1_rhsB = (- a_ + a_14 + a_23 \
      + a_12_34 + a_13_24 - a_14_23)/2

# s_1324, s_1423
aid1_lhsC = a_1324
aid1_rhsC = (- a_ + a_12 + a_34 \
      + a_13_24 + a_14_23 - a_12_34)/2


### for s:
# 3-cycle replacements
sid0_lhsA = s_123
sid0_rhsA = (s_12 + s_23 + s_13 - s_)/2

sid0_lhsB = s_124
sid0_rhsB = (s_12 + s_24 + s_14 - s_)/2

sid0_lhsC = s_134
sid0_rhsC = (s_13 + s_34 + s_14 - s_)/2

sid0_lhsD = s_234
sid0_rhsD = (s_23 + s_34 + s_24 - s_ )/2

#s_1432, s_1234
sid1_lhsA = s_1234
sid1_rhsA = (- s_ + s_13 + s_24 \
      + s_12_34 + s_14_23 - s_13_24)/2

#s_1243, s_1342
sid1_lhsB = s_1243
sid1_rhsB = (- s_ + s_14 + s_23 \
      + s_12_34 + s_13_24 - s_14_23)/2

# s_1324, s_1423
sid1_lhsC = s_1324
sid1_rhsC = (- s_ + s_12 + s_34 \
      + s_13_24 + s_14_23 - s_12_34)/2

def run_substitutions(A):  

    # set elements equal (real Hamiltonian)
    A = A.substitute(a_132 == a_123)
    A = A.substitute(a_142 == a_124)
    A = A.substitute(a_143 == a_134)
    A = A.substitute(a_243 == a_234)
    A = A.substitute(a_1432 == a_1234)
    A = A.substitute(a_1342 == a_1243)
    A = A.substitute(a_1423 == a_1324)

    # set elements equal (real Hamiltonian)
    A = A.substitute(s_132 == s_123)
    A = A.substitute(s_142 == s_124)
    A = A.substitute(s_143 == s_134)
    A = A.substitute(s_243 == s_234)
    A = A.substitute(s_1432 == s_1234)
    A = A.substitute(s_1342 == s_1243)
    A = A.substitute(s_1423 == s_1324)
    
    # apply identities
    # 3-cycles
    A = A.substitute(aid0_lhsA == aid0_rhsA)
    A = A.substitute(aid0_lhsB == aid0_rhsB)
    A = A.substitute(aid0_lhsC == aid0_rhsC)
    A = A.substitute(aid0_lhsD == aid0_rhsD)
    
    # 4-cycles
    A = A.substitute(aid1_lhsA == aid1_rhsA)
    A = A.substitute(aid1_lhsB == aid1_rhsB)
    A = A.substitute(aid1_lhsC == aid1_rhsC)

    # apply identities: for s
    # 3-cycles
    A = A.substitute(sid0_lhsA == sid0_rhsA)
    A = A.substitute(sid0_lhsB == sid0_rhsB)
    A = A.substitute(sid0_lhsC == sid0_rhsC)
    A = A.substitute(sid0_lhsD == sid0_rhsD)
    
    # 4-cycles
    A = A.substitute(sid1_lhsA == sid1_rhsA)
    A = A.substitute(sid1_lhsB == sid1_rhsB)
    A = A.substitute(sid1_lhsC == sid1_rhsC)

    return A
```

```python
def construct_rep(A,lam):
    """ maps A an elment of CS_4 to its representation, lambda a partititon, e.g. [3,1]"""
    
    spc = SymmetricGroupRepresentation(lam, 'orthogonal')
    A_rep = 0
    
    for el in A:
        A_rep = A_rep + el[1]*spc.representation_matrix(Permutation(el[0]))
    return A_rep
```
